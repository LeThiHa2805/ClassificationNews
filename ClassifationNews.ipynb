```python
import json
import csv
import pandas as pd
def file(file_path):
    with open(file_path,'r') as file:
        json_data=json.load(file)

    csv_file_path = 'data.csv'

    # Open the CSV file for writing
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)

        # Write the header row
        header = ['topic', 'url', 'title', 'abstract', 'content']
        csv_writer.writerow(header)

        # Write the data rows
        for topic, articles in json_data.items():
            for article in articles:
                csv_writer.writerow([topic, article['url'], article['title'], article['abstract'], article['content']])
    data=pd.read_csv('/kaggle/working/data.csv')
    return data
```

```python
import re
from pyvi import ViTokenizer
from nltk.corpus import stopwords
import nltk

stopwords_path = '/kaggle/input/stopword/vietnamese-stopwords.txt'
with open(stopwords_path, 'r', encoding='utf-8') as file:
    vietnamese_stopwords = set(file.read().strip().split())

def preprocess_text(text):
    # Chuyển về chữ thường
    text = text.lower()
    
    # Loại bỏ các ký tự không cần thiết, chỉ giữ lại chữ cái và khoảng trắng
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenize và chuẩn hóa từ tiếng Việt sử dụng ViTokenizer từ pyvi
    tokens = ViTokenizer.tokenize(text).split()
    
    # Loại bỏ các từ dừng
    tokens = [word for word in tokens if word not in vietnamese_stopwords]
    
    # Ghép lại các từ thành chuỗi
    processed_text = ' '.join(tokens)
    
    return processed_text

# Áp dụng hàm preprocess_text lên từng phần tử trong X
X_processed = [preprocess_text(text) for text in X]
```

```python
# Khởi tạo TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=10000,  # Số lượng từ vựng tối đa   # Không sử dụng stop words
                                   ngram_range=(1, 2))  # Sử dụng cả unigram và bigram

# Biến đổi dữ liệu văn bản thành ma trận TF-IDF
X_tfidf = tfidf_vectorizer.fit_transform(X_processed)

# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```
